## Examined "Wage or Salary Income in the Past 12 Months" Dataset
- Lists number of people per neighborhood that earn a wage or salary and those that do not
- Provides evidence relating to Well-Being as it relates to financial security/unemployment rate

```python
import csv
import pandas as pd

df = pd.read_csv("Income.csv", sep=",")

df["Proportion with wage or salary"]= df["Estimate; Total: - With wage or salary income"]/df["Estimate; Total:"]

df_sort= df.sort_values("Proportion with wage or salary", ascending=False)

df_sort.iloc[:20].plot.bar(x='Neighborhood', y='Proportion with wage or salary')

```
