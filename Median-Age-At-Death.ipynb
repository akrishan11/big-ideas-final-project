# Median Age At Death

## Examined "Median Age at Death 2011-15 by Neighborhood" Dataset
- Lists median age of death by neighborhood
- Median age of death may correlate to overall health of a population

```python
import csv
import pandas as pd

df = pd.read_csv("Median-Age-At-Death.csv", sep=",")

df_sort= df.sort_values("TOTAL MD AGE AT DEATH", ascending=False)

df_sort.iloc[:20].plot.bar(x='NEIGHBORHOOD', y='TOTAL MD AGE AT DEATH')
```
