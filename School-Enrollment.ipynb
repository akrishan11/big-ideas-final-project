# School Enrollment
## Examined "Pittsburgh American Community Survey 2015, School Enrollment" Dataset
- Lists school enrollment of population by neighborhood from elementary to graduate school
- School enrollment may correlate with future career success of population


```python
import pandas as pd
import os

print(os.getcwd())

df = pd.read_csv("big-ideas-final-project/School-Enrollment.csv", sep=",")

# Adds column which calculates the ratio of children enrolled in school to total children

df["Proportion enrolled"]= df["Estimate; Enrolled in school:"]/df["Estimate; Total:"]

#Sorts neighborhoods by this ratio in descending order and creates a bar graph of the top 15

df_sort= df.sort_values("Proportion enrolled", ascending=False)

df_sort.iloc[:15].plot.bar(x='Neighborhood', y='Proportion enrolled')



```
